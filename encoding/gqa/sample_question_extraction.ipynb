```python
import json 

with open('../../data/gqa/questions/train_balanced_questions.json') as f:
    questions = json.load(f)
```

```python
with open('../../data/gqa/sceneGraphs/train_sceneGraphs.json') as f:
    scene_graphs = json.load(f)
```

```python
import random 

sampled_question_ids = random.sample(list(questions), 100)
sampled_questions = {question_id: questions[question_id] for question_id in sampled_question_ids}

for q_id, q in sampled_questions.items():
    q['sceneGraph'] = scene_graphs[q['imageId']]
```

```python
with open('../../data/gqa/questions/train_sampled_questions.json', 'w') as f:
    json.dump(sampled_questions, f, indent=4)
```

```python
import string 

ontology_mapping = {}
for q_id, q in sampled_questions.items():
    words = q['question'].translate(str.maketrans('', '', string.punctuation)).lower().split(' ')
    print("")
    for word_index, object_id in q['annotations']['question'].items():
        if scene_graphs[q['imageId']]['objects'][object_id]['name'] != words[int(word_index)]:
            ontology_mapping[words[int(word_index)]] = ontology_mapping.get(words[int(word_index)], set())
            ontology_mapping[words[int(word_index)]].add(scene_graphs[q['imageId']]['objects'][object_id]['name'])
```
